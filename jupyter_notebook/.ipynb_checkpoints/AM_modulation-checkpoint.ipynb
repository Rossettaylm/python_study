## AM_modulation

```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
Fs = 10000
Fc = 10
Fw = 1
x = np.linspace(0, 2*pi, Fs)
carrier = np.array([np.sin(2 * pi * Fc * value) for value in x])
modulate_signal = np.array([np.sin(2 * pi * Fw * value) for value in x])
AM_modulation = carrier * modulate_signal
plt.figure(figsize=(10, 6), dpi=128)
plt.plot(x, AM_modulation, linewidth=3, c='pink')
plt.show()
```
